```python
from collections import Counter 
import random 


class MasterMind():
    """
        Jeu MasterMind
        Difficulté : Niveu de difficulté : entre 1 et 3 qui correspond à 3 à 6 couleurs à deviner
        Nombre de tentatives (défaut = 100) : Nombre maximum de tentative
        Couleurs : Choix de couleurs
        Partie facile (défaut = Non): la position exacte est renvoyée 
        
    """
    

    def __init__(self, player_name ='Unknown', tentatives = 100, difficulte = 1, colors = ['jaune', 'rouge', 'vert', 'bleu', 'violet', 'noir'], aide=False):
        self.difficulte = difficulte
        self.player_name = player_name
        self.tentatives = tentatives
        self.colors = colors 
        self.aide = aide
        
        
            
    
    def generer_solution(self):
        if self.difficulte == 1 :
            self.nb_boules = 3
        elif self.difficulte == 2: 
            self.nb_boules = 4
        else :
            self.nb_boules = 5
            
        
        '''Génére une combinaison aléatoire de couleurs '''
        return(random.sample(self.colors, self.nb_boules))
    
  
    
    def jouer(self,jeu):
        """
        Demande au joueur ces propostitions de couleurs . 
        Seul les couleurs valides sont acceptés . 
        
        """
        essais = []
        while len(self.historique) < self.tentatives:
            essais = []
            for count in range(jeu.nb_boules) :
            
                prompt =  f"Entrez votre proposition {str(count + 1)}: "
                essai = input(prompt)
                
                
                while essai not in self.colors:
                    print("Vous devez choisir une couleur dans la liste (sans guillemet): ", jeu.colors)
                    essai = input(prompt)
                    
                else:
                    
                    essais.append(essai)
            self.historique.append(essais)
            resultat = jeu.check_game(essais)
            
            if resultat['correct_positions'] == jeu.nb_boules:
                print(f"Félicitations, vous avez gagné en {len(self.historique)} tentative(s) !")
                break
                
            else :
                if jeu.aide == "True" :
                    print(f"Vous avez trouvé {resultat['correct_colors']} couleurs. Vous avez bien placé {resultat['correct_positions']} couleur(s).")    
                else :
                    print(f"Vous avez trouvé {resultat['correct_positions']} couleurs.")    
            
            self.record_result()
        
        if len(self.historique) >= self.tentatives:
        
            print("Vous avez atteint le nombre d'essais maximum, la solution etait :", jeu.solution)
            
        return(None)
    
    def check_game(self, essai) :
        '''Cette fonction compare la bonne combinaison avec la combinaison proposé par le joueur. 
        
        Retourn : le nbre de couleurs correctement placées et si le nbre de couleurs proposés est correcte (si l'argument self.easy_masterming est oui) facilitant le jeu 
        '''
        correct_positions = 0
        correct_colors = sum(
            (Counter(self.solution) & Counter(essai)).values())
        for i in range(len(essai)):
            if essai[i] == self.solution[i]:
                correct_positions += 1
                
        new_entry = {"correct_colors": correct_colors,
                     "correct_positions": correct_positions}
 
        return(new_entry)

        
    def record_result(self):
        print(self.historique)
        return(self.historique)
    

   
```

```python
   
class Player(MasterMind):
    def __init__(self, player_name ='Unknown', tentatives = 100, difficulte = 1, aide=False, historique=[]):
        super().__init__(player_name)
        super().__init__(difficulte)
        super().__init__(tentatives)
        super().__init__(aide)
        self.historique=historique
    
    def Accueillir(self):
            
        self.player_name = input('Quel est votre pseudo :')
        self.difficulte = int(input('Choisissez la difficulte entre 1 et 3 : '))
        self.tentatives = int(input('Choisissez le nombre de tentatives : '))
        self.aide = input('Voulez vous une partie facile ? (oui ou non)')
        print('\nBonjour', self.player_name)
        print('Bienvenue sur le jeu Master Mind')
        print('\nVoici vos parametres : \nDifficulte :', self.difficulte, '\nNombre de tentatives :', self.tentatives, '\nAide :', self.aide, '\nBoules disponibles :', self.colors)
    
        
        
```
